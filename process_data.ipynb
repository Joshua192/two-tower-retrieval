```python
doc_text_to_id = {}
doc_id_to_text = {}
docID = 0
for doc in docs:
    if doc not in doc_text_to_id:
        doc_text_to_id[doc] = docID
        doc_id_to_text[docID] = doc
        docID+=1 
```

```python
import random
# Built [query_id, pos_doc_id, neg_doc_id] triplets per batch
def build_triplets_batch(batch):
    output = {
        "query_id": [],
        "pos_doc_id": [],
        "neg_doc_id": [],
    }

    for query_id, passage_dict in zip(batch["query_id"], batch["passages"]):
        passage_texts = passage_dict["passage_text"]
        is_selected_flags = passage_dict["is_selected"]
        
        # Get a positive doc
        pos_doc_id = None
        for i, flag in enumerate(is_selected_flags):
            if flag:
                pos_doc_id = doc_text_to_id.get(passage_texts[i])
                break
        if pos_doc_id is None:
            continue

        # Create a set of current query's docs
        current_query_docs = set(passage_texts)

        # Sample a negative doc efficiently
        neg_doc_text = None
        attempts = 0
        while attempts < 10:  # Try a few times to avoid rare edge cases
            candidate = random.choice(docs)
            if candidate not in current_query_docs:
                neg_doc_text = candidate
                break
            attempts += 1
        if neg_doc_text is None:
            continue  # fallback if sampling fails

        neg_doc_id = doc_text_to_id[neg_doc_text]

        output["query_id"].append(query_id)
        output["pos_doc_id"].append(pos_doc_id)
        output["neg_doc_id"].append(neg_doc_id)

    return output

batched_output = ds["train"].map(
    build_triplets_batch,
    batched=True,
    remove_columns=ds["train"].column_names,
    desc="Building triplets" # progress bar
)
triple_id_db = pd.DataFrame(batched_output, columns=["query_id", "pos_doc_id", "neg_doc_id"])
```

```python
import random
from tqdm.auto import tqdm

def map_build_triplets_over_df(
    main: pd.DataFrame,
    build_triplets_batch_fn,
    docs: list,
    doc_text_to_id: dict,
    batch_size: int = 500
) -> pd.DataFrame:
    """
    Apply build_triplets_batch_fn in batches over `main`, return a DataFrame of triplets.

    Parameters
    ----------
    main : pd.DataFrame
        Your input DataFrame with a 'passages' column (dicts with keys 'passage_text' & 'is_selected')
        and a 'query_id' column.
    build_triplets_batch_fn : callable
        The function you supplied; expects a dict-of-lists and returns a dict with keys
        'query_id', 'pos_doc_id', 'neg_doc_id' (each a list).
    docs : list
        The global list of all doc texts used for negative sampling.
    doc_text_to_id : dict
        Mapping from document text to its unique ID.
    batch_size : int, optional (default=500)
        How many rows of `main` to process per batch.

    Returns
    -------
    triple_id_db : pd.DataFrame
        A DataFrame with columns ['query_id', 'pos_doc_id', 'neg_doc_id'].
    """
    outputs = {"query_id": [], "pos_doc_id": [], "neg_doc_id": []}

    # Pre-slice number of batches
    n = len(main)
    n_batches = (n + batch_size - 1) // batch_size

    # Iterate with a tqdm progress bar
    for i in tqdm(range(n_batches), desc="Building triplets"):
        start = i * batch_size
        end = min(start + batch_size, n)
        batch_df = main.iloc[start:end]

        # Convert to HuggingFace-like dict-of-lists
        batch_dict = {
            "query_id": batch_df["query_id"].tolist(),
            "passages": batch_df["passages"].tolist(),
        }

        # Call your provided function
        print("pre-function call")
        out = build_triplets_batch_fn(batch_dict)
        print("func call output:",out,"\n")
        # Accumulate
        outputs["query_id"].extend(out["query_id"])
        outputs["pos_doc_id"].extend(out["pos_doc_id"])
        outputs["neg_doc_id"].extend(out["neg_doc_id"])

    # Build the final DataFrame
    triple_id_db = pd.DataFrame(outputs)
    return triple_id_db
```

```python
triple_id_db = map_build_triplets_over_df(main, build_triplets_batch, docs, doc_text_to_id)
```
